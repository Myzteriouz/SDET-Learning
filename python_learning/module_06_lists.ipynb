# Module 06: Lists

Learn about lists, their properties, indexing, slicing, methods, and how to add, update, and remove elements from lists.

## Creating and Accessing Lists

```python
fruits = ["apple", "banana", "cherry"]
print("Original list:", fruits)
print("Type:", type(fruits))

# Indexing
print("First element:", fruits[0])
print("Last element:", fruits[-1])

# Slicing
print("Slice [0:2]:", fruits[0:2])
```

## Adding Elements

```python
fruits = ["apple", "banana"]
print("Initial:", fruits)

# append()
fruits.append("cherry")
print("After append:", fruits)

# insert()
fruits.insert(1, "blueberry")
print("After insert:", fruits)

# extend()
fruits.extend(["date", "elderberry"])
print("After extend:", fruits)
```

## Updating and Removing Elements

```python
fruits = ["apple", "banana", "cherry", "date"]
print("Initial:", fruits)

# Update
fruits[2] = "blackberry"
print("After update:", fruits)

# remove()
fruits.remove("apple")
print("After remove:", fruits)

# pop()
popped = fruits.pop()
print("Popped:", popped)
print("After pop:", fruits)
```

## List Methods and Operations

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print("Original:", numbers)

# Length
print("Length:", len(numbers))

# Sort
sorted_nums = sorted(numbers)
print("Sorted:", sorted_nums)

# Reverse
print("Reversed:", numbers[::-1])

# Count
print("Count of 1:", numbers.count(1))

# Index
print("Index of 4:", numbers.index(4))
```
