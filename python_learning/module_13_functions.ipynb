Understand functions, how to create and call them, pass arguments, and other essential concepts related to functions.

## Defining and Calling Functions

```python
def greet(name: str) -> str:
    """Return a greeting message for the supplied name."""
    return f"Hello, {name}!"

# Call the function
message = greet("Alice")
print(message)
```

## Parameters and Arguments

```python
def install_package(name: str, version: str = "latest") -> None:
    """Install a package with specified version."""
    print(f"Installing {name} version {version}")

# Positional argument
install_package("requests")

# Keyword arguments
install_package(name="flask", version="2.0.0")
```

## Returning Values

```python
# Return single value
def square(x: int) -> int:
    return x * x

print("Square of 5:", square(5))

# Return multiple values
def min_max(values: list) -> tuple:
    return min(values), max(values)

minimum, maximum = min_max([3, 7, 1, 9])
print(f"Min: {minimum}, Max: {maximum}")
```
