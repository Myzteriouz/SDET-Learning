## Step 2: Variables and Basic Data Types

```python
# Integer
age = 25
print("Age:", age, "Type:", type(age))

# Float
height = 5.8
print("Height:", height, "Type:", type(height))

# Boolean
is_student = True
print("Is student:", is_student, "Type:", type(is_student))

# String
message = "Python is fun"
print("Message:", message, "Type:", type(message))
```

## Step 3: Typecasting

```python
# String to integer
year = "2026"
year_int = int(year)
print("Year as int:", year_int, "Type:", type(year_int))

# Integer to string
age = 25
age_str = str(age)
print("Age as string:", age_str, "Type:", type(age_str))

# String to float
price = "19.99"
price_float = float(price)
print("Price as float:", price_float, "Type:", type(price_float))
```

## Step 4: Multiple Assignment

```python
x, y, z = 1, 2, 3
print("x:", x, "y:", y, "z:", z)

# Swapping values
a, b = 10, 20
a, b = b, a
print("After swap - a:", a, "b:", b)
```
