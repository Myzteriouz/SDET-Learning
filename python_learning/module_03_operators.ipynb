# Module 03: Operators in Python

Understand different operators in Python and their functionalities.

## Arithmetic Operators

```python
x = 10
print("x =", x)
print("Addition (x + 5):", x + 5)
print("Subtraction (x - 3):", x - 3)
print("Multiplication (x * 2):", x * 2)
print("Division (x / 4):", x / 4)
print("Floor Division (x // 3):", x // 3)
print("Remainder (x % 3):", x % 3)
print("Exponent (x ** 2):", x ** 2)
```

## Comparison Operators

```python
x = 10
print("x == 10:", x == 10)
print("x != 5:", x != 5)
print("x > 3:", x > 3)
print("x < 15:", x < 15)
print("x >= 10:", x >= 10)
print("x <= 10:", x <= 10)
```

## Logical Operators

```python
a = True
b = False
print("a and b:", a and b)
print("a or b:", a or b)
print("not a:", not a)
print("not b:", not b)
```

## Assignment Operators

```python
x = 10
print("x =", x)

x += 5
print("After x += 5:", x)

x -= 3
print("After x -= 3:", x)

x *= 2
print("After x *= 2:", x)

x //= 4
print("After x //= 4:", x)
```
