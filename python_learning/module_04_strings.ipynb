# Module 04: Strings

Learn how to create, format, index, and slice strings in Python, along with string methods.

## String Creation and Output

```python
text = "Learning Python is powerful"
print(text)
```

## Indexing and Slicing

```python
text = "Learning Python is powerful"
print("Length:", len(text))

# Indexing
first_char = text[0]
print("First character:", first_char)

# Slicing
print("Slice [0:8]:", text[0:8])
print("Slice [9:15]:", text[9:15])
print("Last 8 characters:", text[-8:])
```

## String Formatting

```python
# f-string formatting
name = "Alice"
score = 92.5
print(f"{name} scored {score}% in Python.")

# format() method
print("{} scored {}%".format(name, score))

# Concatenation
greeting = "Hello, " + name
print(greeting)
```

## String Methods

```python
text = "Learning Python is powerful"

# Case conversion
print("Upper:", text.upper())
print("Lower:", text.lower())
print("Title:", text.title())

# String replacement
print("Replace:", text.replace("powerful", "fun"))

# Split and join
words = text.split()
print("Split:", words)
print("Join:", "-".join(words))

# Search
print("Find 'Python':", text.find("Python"))
print("Starts with 'Learning':", text.startswith("Learning"))
```
