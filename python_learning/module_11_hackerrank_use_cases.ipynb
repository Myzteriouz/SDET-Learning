## What is HackerRank?

- An online platform for competitive programming and skill assessment
- Offers Python challenges for beginners to advanced
- Earn badges as you solve problems

## Step-by-step

1. Register on HackerRank and choose Python challenges.
2. Read the problem carefully and identify input/output formats.
3. Write a small Python script to solve the logic.
4. Test with sample inputs.
5. Submit and iterate until all tests pass.
