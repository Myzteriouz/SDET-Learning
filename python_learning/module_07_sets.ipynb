# Module 07: Sets

Learn about sets, their syntax, operations, methods, and the difference between sets, tuples, and lists.

## Creating Sets

```python
fruits = {"apple", "banana", "cherry"}
print("Set:", fruits)
print("Type:", type(fruits))

# Empty set
empty_set = set()
print("Empty set:", empty_set)
```

## Adding and Removing Elements

```python
fruits = {"apple", "banana", "cherry"}
print("Initial:", fruits)

# Add
fruits.add("date")
print("After add:", fruits)

# Discard
fruits.discard("banana")
print("After discard:", fruits)
```

## Set Operations

```python
odds = {1, 3, 5, 7}
evens = {2, 4, 6, 8}
primes = {2, 3, 5, 7}

print("Odds:", odds)
print("Evens:", evens)
print("Primes:", primes)

# Union
print("\nUnion (odds | evens):", odds | evens)

# Intersection
print("Intersection (odds & primes):", odds & primes)

# Difference
print("Difference (odds - primes):", odds - primes)
```

## Set Properties: Unordered and Unique

```python
# Duplicates are automatically removed
duplicates = {"apple", "apple", "banana", "banana", "cherry"}
print("Remove duplicates:", duplicates)

# Remove duplicates from a list
my_list = [1, 2, 2, 3, 3, 3, 4, 4]
unique_list = list(set(my_list))
print("Unique from list:", unique_list)
```
