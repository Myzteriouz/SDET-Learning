# Module 05: Tuples

Understand tuples, their syntax, properties, indexing, slicing, and methods.

## Creating Tuples

```python
# Create tuples
coords = (10, 20)
print("Tuple:", coords)
print("Type:", type(coords))

# Single-element tuple
single = (42,)
print("Single-element tuple:", single)

# Tuple without parentheses
triple = 1, 2, 3
print("Triple:", triple)
```

## Indexing and Slicing

```python
coords = (10, 20, 30, 40)
print("Tuple:", coords)

# Indexing
print("Index [0]:", coords[0])
print("Index [-1]:", coords[-1])

# Slicing
print("Slice [0:2]:", coords[0:2])
print("Slice [1:]:", coords[1:])
print("Slice [::2]:", coords[::2])
```

## Tuple Methods

```python
values = (3, 1, 4, 1, 5, 9, 2, 6)
print("Tuple:", values)

# count() method
print("Count of 1:", values.count(1))

# index() method
print("Index of 4:", values.index(4))

# Length
print("Length:", len(values))
```

## Immutability

```python
coords = (10, 20)
print("Tuple:", coords)

# Tuples are immutable: the following will fail
try:
    coords[0] = 5
except TypeError as err:
    print("Immutability error:", err)
```

## Use Cases

### Simple
Store fixed coordinate pairs.

### Medium
Return multiple values from a function.

### Difficult
Use tuples as dictionary keys in a mapping.
