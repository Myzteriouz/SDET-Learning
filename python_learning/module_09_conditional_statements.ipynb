# Module 09: Python Conditional Statements

Learn how to set logic with conditional statements and use if, if-else, and if-elif-else statements.

## Simple if Statement

```python
value = 15

if value > 10:
    print("Value is greater than 10")

if value % 2 == 0:
    print("Value is even")
else:
    print("Value is odd")
```

## if-else Statement

```python
age = 20

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

## if-elif-else Statement

```python
value = 0

if value < 0:
    print(f"{value} is negative")
elif value == 0:
    print(f"{value} is zero")
else:
    print(f"{value} is positive")
```

## Nested Conditions

```python
value = 15

if value > 0:
    if value < 20:
        print("Value is positive and less than 20")
    else:
        print("Value is positive and >= 20")
else:
    print("Value is not positive")
```

## Comparison and Logical Operators

```python
age = 25
has_license = True

# Logical AND
if age >= 18 and has_license:
    print("You can drive")

# Logical OR
score = 45
if score >= 90 or score >= 80:
    print("Good score")

# Logical NOT
if not (score < 50):
    print("Score is acceptable")
```
