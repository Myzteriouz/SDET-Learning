# Module 10: Loops in Python

Understand loops in Python, including while and for loops, range, break, continue, pass, enumerate, zip, and assert.

## While Loop

```python
count = 1
while count <= 5:
    print("While loop:", count)
    count += 1
```

## For Loop with range()

```python
# Basic for loop
for i in range(5):
    print("For loop:", i)
```

## Range with Step

```python
# range(start, stop, step)
for i in range(0, 10, 2):
    print("Step:", i)
```

## Break and Continue

```python
print("Break example:")
for i in range(5):
    if i == 3:
        break
    print(i)

print("\nContinue example:")
for i in range(5):
    if i == 2:
        continue
    print(i)
```

## Enumerate

```python
colors = ["red", "green", "blue"]
for index, color in enumerate(colors, start=1):
    print(f"{index}: {color}")
```

## Zip

```python
numbers = [1, 2, 3]
words = ["one", "two", "three"]
for number, word in zip(numbers, words):
    print(f"{number} => {word}")
```

## Assert

```python
colors = ["red", "green", "blue"]
assert len(colors) == 3, "Expected exactly three colors"
print("Assertion passed")
```
